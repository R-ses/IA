```python
import numpy as np
import matplotlib
import matplotlib.pyplot as ml
import math as mt
```

```python
class Trimf():
    def __init__(self,a,m,b):
        self.a = a
        self.m = m
        self.b = b
    def f_eval(self,X):        
        Y = []
        for i in X:
            if i<self.a:
                Y.append(0)
            else:
                if self.a<=i and i<=self.m:
                    Y.append((i-self.a)/(self.m-self.a))
                else:
                    if self.m<=i and i<=self.b:
                        Y.append((self.b-i)/(self.b-self.m))
                    else:
                        if(self.b<i):
                            Y.append(0)
                        else:
                            if(i==self.m):
                                Y.append(1)
        
        return Y
```

```python
class Trapmf():
    def __init__(self,a,b,c,d):
        self.a=a
        self.b=b
        self.c=c
        self.d=d
        
    def f_eval(self,X):
        Y = []
        for i in X:
            if i < self.a or i > self.d:
                Y.append(0)
            else:
                if self.a <= i and i <= self.b: 
                    Y.append((i-self.a)/(self.b-self.a))
                else:
                    if self.b<= i and i <= self.c:
                        Y.append(1)
                    else:
                        if self.c <= i <= self.d:
                            Y.append((self.d-i)/(self.d-self.c))
        return Y
```

```python
class Gbellmf():
    def __init__(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c
    def f_eval(self,X):
        Y = []
        for i in X:
            Y.append((1)/(1+(abs((i-self.c)/(self.a)))**(2*self.b)))
        return Y
    
```

```python
class Gaussmf():
    def __init__(self,a,b):
        self.a=a
        self.b=b
        
    def f_eval(self,X):
        Y = []
        for i in X:
            Y.append(mt.exp((-(i-self.a)**2)/(2*(self.b**2))))
        return Y
```

```python
X = range(0,20)

A = Trimf(0,5,10)
B = Trapmf(0,2,8,10)
C = Gbellmf(3,5,8)
D = Gaussmf(5,10)

A.f_eval(X)
B.f_eval(X)
C.f_eval(X)
D.f_eval(X)

ml.plot(X,A.f_eval(X))
ml.figure()
ml.plot(X,B.f_eval(X))
ml.figure()
ml.plot(X,C.f_eval(X))
ml.figure()
ml.plot(X,D.f_eval(X))
```
