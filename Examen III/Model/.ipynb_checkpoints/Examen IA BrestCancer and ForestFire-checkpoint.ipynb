# Funcion extra

```python
def get_model():
    model = Sequential()
    #Capa de entrada y oculta
    model.add(Dense(10, input_shape=(9,), kernel_initializer='random_uniform'))
    #CAPAS
    model.add(Dense(2, activation ='relu'))
    #Capa de salida
    #Neuronas de salida igual a la cantidad de etiquetas
    model.add(Dense(2))
    model.add(Activation('softmax'))
    
    #Método de aprendizaje
    #opt = SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)#Stochastic gradient descent 
    opt = Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)#ADAM
    #opt = Adagrad(lr=0.01, epsilon=None, decay=0.0)
    #opt = Adadelta(lr=1.0, rho=0.95, epsilon=None, decay=0.0)
    #opt = Adamax(lr=0.002, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0)
    #opt = Nadam(lr=0.002, beta_1=0.9, beta_2=0.999, epsilon=None, schedule_decay=0.004)
    

    model.compile(optimizer=opt,
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model
```

```python
%matplotlib notebook

def view3D(X, y, model = decomposition.PCA):
    fig = plt.figure(1,figsize=(8, 6))
    ax = Axes3D(fig, elev=-150, azim=110)
    X_reduced = model(n_components=3).fit_transform(X)
    ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=y,
             cmap=plt.cm.Set1, edgecolor='k', s=40)
    ax.set_title("First three PCA directions")
    ax.set_xlabel("1st eigenvector")
    ax.w_xaxis.set_ticklabels([])
    ax.set_ylabel("2nd eigenvector")
    ax.w_yaxis.set_ticklabels([])
    ax.set_zlabel("3rd eigenvector")
    ax.w_zaxis.set_ticklabels([])
    plt.show()
    
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
```

```python
dataset1 = pd.read_csv('../Dataset/WBCD.csv')
```

# Descripcion de Dataset Wisconsin Breast Cancer 

Ejercicio de Classificacion WBCD el dataset consta de 11 atributos incluyendo el ID de la muestra, en el siguiente orden   
   -- -----------------------------------------
   1. Sample code number            id number
   2. Clump Thickness               1 - 10
   3. Uniformity of Cell Size       1 - 10
   4. Uniformity of Cell Shape      1 - 10
   5. Marginal Adhesion             1 - 10
   6. Single Epithelial Cell Size   1 - 10
   7. Bare Nuclei                   1 - 10
   8. Bland Chromatin               1 - 10
   9. Normal Nucleoli               1 - 10
  10. Mitoses                       1 - 10
  11. Class:                        (2 for benign, 4 for malignant)
  

## Preprocessing

### Remove "missing data"

```python
dataset1 = dataset1.replace('?',0)
```

```python
X1 = dataset1.drop(columns=["Class","Id_sample"])
y1 = dataset1["Class"]
```

## Reduccion de dimensionalidad
###  Transformacion de datos

```python

SVD = TruncatedSVD(n_components=3, n_iter=5)
RD_datasetX = SVD.fit_transform(X1)
```

#### Clasificacion Modelo 1:
###### SGDClassifier

```python
#from sklearn.linear_model import SGDClassifier

k_fold = KFold(n_splits = 4)
SGDC = linear_model.SGDClassifier(loss='perceptron',max_iter=1000, tol=1e-3 )

for k,(train, test) in enumerate (k_fold.split(RD_datasetX)):
    #print(test)
    #print(train)
    # Se entrena con el conjunto de datos de entrenamiento
    #X1[train]
    SGDC.fit(RD_datasetX[train], y1[train])
    # Se valida con el conjunto de datos de prueba
    #print("[fold {}]  coeficiente de correlación: {}".format(k, SGDC.score(RD_datasetX[test], y1[test])))
    
    y1_pred = SGDC.predict(RD_datasetX[test])

    cnf_matrix = confusion_matrix(y1[test], y1_pred)
    class_names = ['benigno','maligno']
    np.set_printoptions(precision=2)

    plt.figure()
    plot_confusion_matrix(cnf_matrix, classes=class_names,
                          title='Matriz de confusion Dataset1 WinsconsinBreastCancer SGDClassifier ')

```

#### Clasificacion Modelo 2:
###### Logistic Regression

```python

X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size=0.7)


reg_log = LogisticRegression(multi_class = 'ovr')
reg_log.fit(X_train, y_train)
y_pred = reg_log.predict(X_test)

cnf_matrix = confusion_matrix(y_test, y_pred)
class_names = ['Benigno','Maligno']
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=class_names,
                      title='Matriz de confusion Dataset1 WinsconsinBreastCancer Logistic Regression')
```

#### Clasificacion Modelo 3:
###### Red Neuronal

```python
kf = StratifiedKFold(n_splits=4)
#print(X1)

for idx_train, idx_test in kf.split(X1,y1):
    X_train = X1.loc[idx_train]
    X_test = X1.loc[idx_test]
    y_train = y1.loc[idx_train]
    y_test = y1.loc[idx_test]  
    
    scaler = StandardScaler()
    scaler.fit(X_train)
    X_train = scaler.transform(X_train)
    X_test = scaler.transform(X_test)

    model = get_model() 
    model.fit(X_train, pd.get_dummies(y_train), epochs=100, verbose=False)
    #print(X_train)
    #print(y_train)
    
    
    y_pred = reg_log.predict(X_test)
cnf_matrix = confusion_matrix(y_test, y_pred)
class_names = ['Benigno','Maligno']
np.set_printoptions(precision=2)

plt.figure()
plot_confusion_matrix(cnf_matrix, classes=class_names,
title='Neural Network Classifier')
```
