# Funcion extra

```python
%matplotlib notebook

def view3D(X, y, model = decomposition.PCA):
    fig = plt.figure(1,figsize=(8, 6))
    ax = Axes3D(fig, elev=-150, azim=110)
    X_reduced = model(n_components=3).fit_transform(X)
    ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=y,
             cmap=plt.cm.Set1, edgecolor='k', s=40)
    ax.set_title("First three PCA directions")
    ax.set_xlabel("1st eigenvector")
    ax.w_xaxis.set_ticklabels([])
    ax.set_ylabel("2nd eigenvector")
    ax.w_yaxis.set_ticklabels([])
    ax.set_zlabel("3rd eigenvector")
    ax.w_zaxis.set_ticklabels([])
    plt.show()
    
```

## Preprocessing

### Dataset 2 winsonsin 
#### Remove "missing data"

```python
#Search into Label enconder _?_
dataset1 = dataset1.replace('?',0)


X1 = dataset1.drop(columns = ['Class'])
y1 = dataset1['Class']
```

## Reduccion de dimensionalidad
###  Transformacion de datos

```python
from sklearn.decomposition import TruncatedSVD
SVD = TruncatedSVD(n_components=5, n_iter=5)
RD_dataset1 = SVD.fit_transform(dataset1)
#Reduccion de dimensiones de 10 a 5
```

#### Clasificacion Modelo 1:
###### SGDClassifier

```python
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import KFold

k_fold = KFold(n_splits = 4)
SGD = SGDClassifier(loss = "hinge", penalty="l2")

for k, (train, test) in enumerate(k_fold.split(X1, y1)):
    # Se entrena con el conjunto de datos de entrenamiento
    SGD.fit(X1[train], y1[train])
    # Se valida con el conjunto de datos de prueba
    print("[fold {}]  coeficiente de correlacin: {}".format(k, SGD.score(X1[test], y1[test])))
```
